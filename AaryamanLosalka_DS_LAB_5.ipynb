```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm, bernoulli, binom, poisson, uniform
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# --- Configuration ---
sns.set_style("whitegrid")
# Setting a standard figure size for consistent plotting
plt.rcParams['figure.figsize'] = [10, 6]
plt.rcParams['figure.dpi'] = 100

# Function to safely print tables in Markdown format
def print_table(df, title):
    """Prints a pandas DataFrame as a Markdown table with a title."""
    print(f"\n--- Calculation/Summary for: {title} ---")
    print(df.to_markdown(numalign="left", stralign="left"))

# ==============================================================================
# SECTION I: Data Loading and Preparation
# ==============================================================================
print("="*80)
print("--- SECTION I: Data Loading and Preparation ---")
print("="*80)

try:
    df = pd.read_csv('Food_Delivery_Times.csv')
except FileNotFoundError:
    # Use placeholder data if file is not found to ensure script execution
    print("Error: 'Food_Delivery_Times.csv' not found. Using placeholder data.")
    df = pd.DataFrame({
        'Delivery_Time_min': [45, 80, 50, 60, 40],
        'Distance_km': [10, 20, 5, 15, 8],
        'Courier_Experience_yrs': [2, 5, 1, 3, 4],
        'Vehicle_Type': ['Bike', 'Car', 'Scooter', 'Bike', 'Car'],
        'Traffic_Level': ['Medium', 'High', 'Low', 'Medium', 'Low'],
        'Time_of_Day': ['Afternoon', 'Evening', 'Morning', 'Afternoon', 'Morning']
    })

# Data Cleaning
df['Delivery_Time_min'] = pd.to_numeric(df['Delivery_Time_min'], errors='coerce')
df['Distance_km'] = pd.to_numeric(df['Distance_km'], errors='coerce')
df['Courier_Experience_yrs'] = pd.to_numeric(df['Courier_Experience_yrs'], errors='coerce')
df_cleaned = df.dropna(subset=['Delivery_Time_min', 'Distance_km', 'Courier_Experience_yrs']).copy()
for col in ['Vehicle_Type', 'Traffic_Level', 'Time_of_Day']:
    df_cleaned[col] = df_cleaned[col].astype(str).replace('nan', 'Unknown')
    df_cleaned[col] = df_cleaned[col].astype('category').cat.remove_unused_categories()

# Define column mapping based on data inspection
COL_MAP = {
    'Delivery_Time_min': 'Delivery_Time_min',
    'Distance_km': 'Distance_km',
    'Courier_Experience_yrs': 'Courier_Experience_yrs',
    'Vehicle_Type': 'Vehicle_Type',
    'Traffic_Level': 'Traffic_Level',
    'Time_of_Day': 'Time_of_Day'
}

print(f"Data loaded and cleaned. {len(df_cleaned)} valid rows remaining for analysis.")

# ==============================================================================
# SECTION II: PROBABILITY DISTRIBUTION TASKS (Simulations & Calculations)
# ==============================================================================
print("\n" + "="*80)
print("--- SECTION II: PROBABILITY DISTRIBUTION TASKS (Simulations & Calculations) ---")
print("="*80)

# 1. Normal Distribution (μ=0, σ=1)
print("\n*** Plot 1: Standard Normal Distribution ($\mu=0, \sigma=1$) ***")
np.random.seed(42)
data_standard_normal = norm.rvs(loc=0, scale=1, size=10000)
norm_calc = pd.DataFrame({
    'Parameter': ['Theoretical Mean (μ)', 'Theoretical Std. Dev (σ)', 'Simulated Mean', 'Simulated Std. Dev', 'Sample Size'],
    'Value': [0, 1, np.mean(data_standard_normal).round(4), np.std(data_standard_normal).round(4), 10000]
}).set_index('Parameter')
print_table(norm_calc, "Standard Normal Distribution Parameters & Simulation Results")
plt.figure(figsize=(10, 6))
sns.histplot(data_standard_normal, bins=50, kde=True, color='purple', stat='density')
plt.title('1. Standard Normal Distribution ($\mu=0, \sigma=1$) Simulation')
plt.xlabel('Value (z-score)')
plt.ylabel('Density')
plt.show()

# 2. Bernoulli Distribution (p=0.3)
print("\n*** Plot 2: Bernoulli Distribution ($p=0.3$) ***")
p_bernoulli = 0.3
np.random.seed(42)
data_bernoulli = bernoulli.rvs(p=p_bernoulli, size=10000)
proportion_clicked = np.sum(data_bernoulli) / len(data_bernoulli)
bern_calc = pd.DataFrame({
    'Metric': ['Theoretical Probability (p)', 'Theoretical Mean (p)', 'Simulated Mean/Proportion', 'Sample Size (N)'],
    'Value': [p_bernoulli, p_bernoulli, f'{proportion_clicked:.4f}', 10000]
}).set_index('Metric')
print_table(bern_calc, "Bernoulli Distribution Parameters & Simulation Results")
plt.figure(figsize=(6, 4))
sns.histplot(data_bernoulli, bins=[-0.5, 0.5, 1.5], stat='count', discrete=True, color='salmon', edgecolor='black', shrink=0.5)
plt.title(f'2. Bernoulli Distribution Simulation ($p={p_bernoulli}$)')
plt.xlabel('Outcome (0=Failure, 1=Success)')
plt.xticks([0, 1])
plt.ylabel('Count')
plt.show()

# 3. Binomial Distribution (n=10, p=0.5)
print("\n*** Plot 3: Binomial Distribution ($n=10, p=0.5$) ***")
n_binom = 10
p_binom = 0.5
np.random.seed(42)
data_binomial = binom.rvs(n=n_binom, p=p_binom, size=10000)
mode_binomial = np.argmax(np.bincount(data_binomial)) 
binom_calc = pd.DataFrame({
    'Metric': ['Number of Trials (n)', 'Success Probability (p)', 'Theoretical Expected Value E[X] = n*p', 'Theoretical Variance Var[X] = n*p*(1-p)', 'Simulated Mode'],
    'Value': [n_binom, p_binom, n_binom * p_binom, n_binom * p_binom * (1 - p_binom), mode_binomial]
}).set_index('Metric')
print_table(binom_calc, "Binomial Distribution Parameters & Simulation Results")
plt.figure(figsize=(8, 5))
sns.histplot(data_binomial, bins=np.arange(-0.5, n_binom + 1.5, 1), stat='count', discrete=True, color='teal', edgecolor='black')
plt.title(f'3. Binomial Distribution Simulation ($n={n_binom}, p={p_binom}$)')
plt.xlabel('Number of Successes (k)')
plt.ylabel('Count')
plt.xticks(range(n_binom + 1))
plt.show()

# 4. Poisson distribution (mu=0.3)
print("\n*** Plot 4: Poisson Distribution ($\mu=0.3$) ***")
mu_poisson = 0.3
np.random.seed(42)
data_poisson = poisson.rvs(mu=mu_poisson, size=10000)
poiss_calc = pd.DataFrame({
    'Metric': ['Theoretical Mean Rate ($\mu$)', 'Theoretical Variance ($\sigma^2$) = $\mu$', 'Simulated Mean', 'Sample Size'],
    'Value': [mu_poisson, mu_poisson, np.mean(data_poisson).round(4), 10000]
}).set_index('Metric')
print_table(poiss_calc, "Poisson Distribution Parameters & Simulation Results ($\mu=0.3$)")
plt.figure(figsize=(8, 5))
sns.histplot(data_poisson, bins=np.arange(-0.5, max(data_poisson) + 1.5, 1), discrete=True, color='olive', edgecolor='black')
plt.title(f'4. Poisson Distribution Simulation ($\mu={mu_poisson}$)')
plt.xlabel('Number of Events (k)')
plt.ylabel('Count')
plt.xticks(np.unique(data_poisson))
plt.show()

# 5. Discrete Uniform Distribution (Die Roll)
print("\n*** Plot 5: Discrete Uniform Distribution (Die Roll) ***")
sides = 6
probabilities = np.full(sides, 1/sides)
expected_value = (sides + 1) / 2
variance = (sides**2 - 1) / 12
uniform_discrete_calc = pd.DataFrame({
    'Metric': ['Number of Outcomes (n)', 'Probability P(X=x)', 'Expected Value E[X]', 'Variance Var[X]'],
    # FIX: Using round() function instead of .round() method on float
    'Value': [sides, f'$1/6 \\approx {probabilities[0]:.4f}$', expected_value, round(variance, 4)] 
}).set_index('Metric')
print_table(uniform_discrete_calc, "Parameters for Discrete Uniform Distribution (Die Roll)")
outcomes = np.arange(1, sides + 1)
plt.figure(figsize=(6, 4))
plt.bar(outcomes, probabilities, color='skyblue', edgecolor='black')
plt.title('5. Discrete Uniform Distribution (Fair Die Roll)')
plt.xlabel('Outcome (x)')
plt.ylabel('Probability P(X=x)')
plt.xticks(outcomes)
plt.ylim(0, 0.25)
plt.grid(axis='y', linestyle='--')
plt.show()

# 6. Continuous Uniform Distribution (U(0, 1))
print("\n*** Plot 6: Continuous Uniform Distribution U(0, 1) ***")
a_cont, b_cont = 0, 1
np.random.seed(42)
data_cont_uniform = uniform.rvs(loc=a_cont, scale=b_cont - a_cont, size=1000000)
# Calculate variance before formatting
theoretical_variance = (b_cont - a_cont)**2 / 12
uniform_cont_calc = pd.DataFrame({
    'Metric': ['Lower Bound (a)', 'Upper Bound (b)', 'PDF Value f(x) = 1/(b-a)', 'Theoretical Mean E[X] = (a+b)/2', 'Theoretical Variance Var[X] = (b-a)^2/12', 'Simulated Mean'],
    # FIX: Using round() function instead of .round() method on float
    'Value': [a_cont, b_cont, 1/(b_cont - a_cont), (a_cont + b_cont) / 2, round(theoretical_variance, 4), np.mean(data_cont_uniform).round(4)]
}).set_index('Metric')
print_table(uniform_cont_calc, "Parameters for Continuous Uniform Distribution U(0, 1)")
plt.figure(figsize=(10, 6))
sns.histplot(data_cont_uniform, bins=100, kde=False, stat='density', color='gold', edgecolor='black')
plt.title('6. Continuous Uniform Distribution U(0, 1) Simulation')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# 7. Poisson distribution (μ=3, Call Center)
print("\n*** Plot 7: Poisson Distribution ($\mu=3$) ***")
mu_poisson_task6 = 3
np.random.seed(42)
data_poisson_task6 = poisson.rvs(mu=mu_poisson_task6, size=10000)
mode_poisson_task6 = np.argmax(np.bincount(data_poisson_task6))
poiss_calc_task6 = pd.DataFrame({
    'Metric': ['Theoretical Mean Rate ($\mu$)', 'Theoretical Mode (Floor($\mu$))', 'Simulated Mode', 'Simulated Mean'],
    'Value': [mu_poisson_task6, f'Floor({mu_poisson_task6}) = {int(mu_poisson_task6)}', mode_poisson_task6, np.mean(data_poisson_task6).round(4)]
}).set_index('Metric')
print_table(poiss_calc_task6, "Calculations for Poisson Distribution ($\mu=3$)")
plt.figure(figsize=(8, 5))
sns.histplot(data_poisson_task6, bins=np.arange(-0.5, max(data_poisson_task6) + 1.5, 1), discrete=True, color='crimson', edgecolor='black')
plt.title(f'7. Poisson Distribution Simulation (μ={mu_poisson_task6})')
plt.xlabel('Number of Calls per Hour')
plt.ylabel('Count')
plt.show()

# ==============================================================================
# SECTION III: VISUALIZATION TASKS (Data Analysis & Plotting)
# ==============================================================================
print("\n" + "="*80)
print("--- SECTION III: VISUALIZATION TASKS (Data Analysis & Plotting) ---")
print("="*80)

numeric_cols = [COL_MAP['Delivery_Time_min'], COL_MAP['Distance_km'], COL_MAP['Courier_Experience_yrs']]

# --- Descriptive Statistics Table (General Calculation) ---
desc_stats = df_cleaned[numeric_cols].describe().T
desc_stats['CoV'] = desc_stats['std'] / desc_stats['mean'] # Coefficient of Variation
desc_stats = desc_stats.round(2)
desc_stats.columns.name = 'Statistic'
print_table(desc_stats, "Overall Descriptive Statistics for Numeric Variables")

# 0a. Pair Plot
print("\n*** Plot 0a: Pair Plot: Delivery Metrics ***")
print_table(desc_stats[['mean', 'std', 'min', 'max']], "Data Summary for Pair Plot")
print("Calculation: None. Visualizes pairwise scatter plots and marginal distributions (KDE) of the three numeric variables.")
sns.pairplot(df_cleaned, vars=numeric_cols, corner=True, diag_kind='kde')
plt.suptitle('0a. Pair Plot: Delivery Time, Distance, and Courier Experience', y=1.02)
plt.show()

# 0b. Clustermap
print("\n*** Plot 0b: Clustermap: Correlation of Delivery Metrics ***")
corr_matrix = df_cleaned[numeric_cols].corr()
print_table(corr_matrix.round(3), "Correlation Matrix (Clustermap Data)")
sns.clustermap(corr_matrix, annot=True, cmap='vlag', linewidths=.5, figsize=(8, 8))
plt.suptitle('0b. Clustermap: Correlation of Delivery Metrics', y=1.02)
plt.show()

# 1. Line Plot (Delivery_Time_min vs. Distance_km)
print("\n*** Plot 1: Line Plot: Delivery Time vs. Distance ***")
line_data = df_cleaned[[COL_MAP['Distance_km'], COL_MAP['Delivery_Time_min']]]
line_summary = line_data.describe().T.round(2)
line_corr = line_data.corr().iloc[0, 1]
line_summary['Correlation'] = [line_corr.round(3), line_corr.round(3)]
print_table(line_summary, "Summary Statistics and Correlation for Line Plot")
df_sorted = df_cleaned.sort_values(by=COL_MAP['Distance_km']).reset_index(drop=True) 
sns.lineplot(x=COL_MAP['Distance_km'], y=COL_MAP['Delivery_Time_min'], data=df_sorted)
plt.title('1. Line Plot: Delivery Time vs. Distance (Sorted by Distance)')
plt.xlabel('Distance (km)')
plt.ylabel('Delivery Time (min)')
plt.show()

# 2. Scatter Plot (Delivery_Time_min vs. Distance_km)
print("\n*** Plot 2: Scatter Plot: Delivery Time vs. Distance ***")
print_table(line_summary, "Summary Statistics and Correlation for Scatter Plot")
plt.figure(figsize=(10, 6))
sns.scatterplot(x=COL_MAP['Distance_km'], y=COL_MAP['Delivery_Time_min'], data=df_cleaned, alpha=0.6)
plt.title('2. Scatter Plot: Delivery Time vs. Distance')
plt.xlabel('Distance (km)')
plt.ylabel('Delivery Time (min)')
plt.show()

# 3. Linear Regression Plot (Delivery_Time_min vs. Courier_Experience_yrs)
print("\n*** Plot 3: Linear Regression: Delivery Time vs. Courier Experience ***")
X = df_cleaned[[COL_MAP['Courier_Experience_yrs']]]
y = df_cleaned[COL_MAP['Delivery_Time_min']]
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
r_squared = r2_score(y, y_pred)
corr = X[COL_MAP['Courier_Experience_yrs']].corr(y).round(4)
reg_summary = pd.DataFrame({
    'Metric': ['Intercept ($\beta_0$)', 'Coefficient ($\beta_1$)', 'R-squared ($R^2$)', 'Pearson Correlation (r)'],
    # FIX: Using round() function instead of .round() method on float
    'Value': [round(model.intercept_, 4), round(model.coef_[0], 4), round(r_squared, 4), corr]
}).set_index('Metric')
print_table(reg_summary, "Linear Regression Model Metrics")
plt.figure(figsize=(10, 6))
sns.regplot(x=COL_MAP['Courier_Experience_yrs'], y=COL_MAP['Delivery_Time_min'], data=df_cleaned, scatter_kws={'alpha':0.4})
plt.title(f'3. Linear Regression: Delivery Time vs. Courier Experience ($R^2$={r_squared:.4f})')
plt.xlabel('Courier Experience (yrs)')
plt.ylabel('Delivery Time (min)')
plt.show()

# 4. Bar Plot (Count by Vehicle_Type)
print("\n*** Plot 4: Bar Plot: Count of Orders by Vehicle Type ***")
counts_vehicle = df_cleaned[COL_MAP['Vehicle_Type']].value_counts().reset_index()
counts_vehicle.columns = ['Vehicle_Type', 'Order_Count']
counts_vehicle['Proportion'] = (counts_vehicle['Order_Count'] / counts_vehicle['Order_Count'].sum()).round(4)
print_table(counts_vehicle, "Counts and Proportions by Vehicle Type")
plt.figure(figsize=(10, 6))
sns.countplot(y=COL_MAP['Vehicle_Type'], data=df_cleaned, order=counts_vehicle['Vehicle_Type'])
plt.title('4. Bar Plot: Count of Orders by Vehicle Type')
plt.xlabel('Count of Orders')
plt.ylabel('Vehicle Type')
plt.show()

# 5. Bar Plot (Count by Traffic_Level)
print("\n*** Plot 5: Bar Plot: Count of Orders by Traffic Level ***")
counts_traffic = df_cleaned[COL_MAP['Traffic_Level']].value_counts().reset_index()
counts_traffic.columns = ['Traffic_Level', 'Order_Count']
counts_traffic['Proportion'] = (counts_traffic['Order_Count'] / counts_traffic['Order_Count'].sum()).round(4)
print_table(counts_traffic, "Counts and Proportions by Traffic Level")
plt.figure(figsize=(10, 6))
sns.countplot(y=COL_MAP['Traffic_Level'], data=df_cleaned, order=counts_traffic['Traffic_Level'])
plt.title('5. Bar Plot: Count of Orders by Traffic Level')
plt.xlabel('Count of Orders')
plt.ylabel('Traffic Level')
plt.show()

# 6. Bar Plot (Average Delivery_Time_min by Vehicle_Type)
print("\n*** Plot 6: Bar Plot: Average Delivery Time by Vehicle Type ***")
avg_delivery_time = df_cleaned.groupby(COL_MAP['Vehicle_Type'])[COL_MAP['Delivery_Time_min']].agg(['count', 'mean', 'std', 'median']).sort_values(by='mean', ascending=False).round(2).reset_index()
avg_delivery_time.columns = ['Vehicle_Type', 'N', 'Mean_Delivery_Time_min', 'Std_Dev', 'Median_Delivery_Time_min']
print_table(avg_delivery_time, "Grouped Statistics for Average Delivery Time by Vehicle Type")
plt.figure(figsize=(10, 6))
sns.barplot(x='Mean_Delivery_Time_min', y='Vehicle_Type', data=avg_delivery_time)
plt.title('6. Bar Plot: Average Delivery Time by Vehicle Type')
plt.xlabel('Average Delivery Time (min)')
plt.ylabel('Vehicle Type')
plt.show()

# 7. Box Plot (Delivery_Time_min Distribution by Time_of_Day)
print("\n*** Plot 7: Box Plot: Delivery Time Distribution by Time of Day ***")
boxplot_data = df_cleaned.groupby(COL_MAP['Time_of_Day'])[COL_MAP['Delivery_Time_min']].agg(['count', 'mean', 'median', lambda x: x.quantile(0.25), lambda x: x.quantile(0.75), 'std'])
boxplot_data.columns = ['N', 'Mean', 'Median', 'Q1 (25th Pctl)', 'Q3 (75th Pctl)', 'Std_Dev']
print_table(boxplot_data.round(2), "Detailed Statistics (Median, IQR, Mean) by Time of Day")
plt.figure(figsize=(12, 7))
sns.boxplot(x=COL_MAP['Time_of_Day'], y=COL_MAP['Delivery_Time_min'], data=df_cleaned)
plt.title('7. Box Plot: Delivery Time Distribution by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Delivery Time (min)')
plt.show()

# 8. Swarm Plot (Courier_Experience_yrs vs. Traffic_Level)
print("\n*** Plot 8: Swarm Plot: Courier Experience vs. Traffic Level ***")
swarm_data = df_cleaned.groupby(COL_MAP['Traffic_Level'])[COL_MAP['Courier_Experience_yrs']].agg(['count', 'mean', 'median', 'std']).reset_index()
swarm_data.columns = ['Traffic_Level', 'N', 'Mean_Experience', 'Median_Experience', 'Std_Dev_Experience']
print_table(swarm_data.round(2), "Grouped Statistics for Courier Experience by Traffic Level (Swarm Plot Data)")
plt.figure(figsize=(12, 7))
sns.swarmplot(x=COL_MAP['Traffic_Level'], y=COL_MAP['Courier_Experience_yrs'], data=df_cleaned, size=3, alpha=0.7)
plt.title('8. Swarm Plot: Courier Experience vs. Traffic Level')
plt.xlabel('Traffic Level')
plt.ylabel('Courier Experience (yrs)')
plt.show()

# 9. Violin Plot (Courier_Experience_yrs Distribution vs. Traffic_Level)
print("\n*** Plot 9: Violin Plot: Courier Experience Distribution vs. Traffic Level ***")
print_table(swarm_data.round(2), "Grouped Statistics for Courier Experience by Traffic Level (Violin Plot Data)")
plt.figure(figsize=(12, 7))
sns.violinplot(x=COL_MAP['Traffic_Level'], y=COL_MAP['Courier_Experience_yrs'], data=df_cleaned)
plt.title('9. Violin Plot: Courier Experience Distribution vs. Traffic Level')
plt.xlabel('Traffic Level')
plt.ylabel('Courier Experience (yrs)')
plt.show()

# 10. Subplots (Swarm Plot and Violin Plot)
print("\n*** Plot 10: Subplots: Swarm Plot and Violin Plot Combined ***")
print_table(swarm_data.round(2), "Grouped Statistics for Combined Subplots (Swarm/Violin)")
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

sns.swarmplot(x=COL_MAP['Traffic_Level'], y=COL_MAP['Courier_Experience_yrs'], data=df_cleaned, size=3, alpha=0.7, ax=axes[0])
axes[0].set_title('10a. Swarm Plot: Courier Experience vs. Traffic Level')
axes[0].set_xlabel('Traffic Level')
axes[0].set_ylabel('Courier Experience (yrs)')

sns.violinplot(x=COL_MAP['Traffic_Level'], y=COL_MAP['Courier_Experience_yrs'], data=df_cleaned, ax=axes[1])
axes[1].set_title('10b. Violin Plot: Courier Experience Distribution vs. Traffic Level')
axes[1].set_xlabel('Traffic Level')
axes[1].set_ylabel('Courier Experience (yrs)')

plt.tight_layout()
plt.show()

print("\n--- Script Execution Complete ---")
```
